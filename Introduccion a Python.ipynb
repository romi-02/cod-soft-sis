## Diccionario

Los diccionarios en Python son una estructura de datos que permite almacenar su contenido en forma de llave y valor.
Los diccionarios se pueden crear con parntesis {} separando con una coma cada par key: value. En el siguiente ejemplo tenemos tres keys que son el nombre, la edad y el documento.

```python
d1 = {
  "Nombre": "Sara",
  "Edad": 27,
  "Documento": 1003882
}
print(d1)
#{'Nombre': 'Sara', 'Edad': 27, 'Documento': 1003882}
```

Otra forma equivalente de crear un diccionario en Python es usando dict() e introduciendo los pares key: value entre parntesis.

```python
d2 = dict([
      ('Nombre', 'Sara'),
      ('Edad', 27),
      ('Documento', 1003882),
])
print(d2)
#{'Nombre': 'Sara', 'Edad': '27', 'Documento': '1003882'}
```

Tambin es posible usar el constructor dict() para crear un diccionario.

```python
d3 = dict(Nombre='Sara',
          Edad=27,
          Documento=1003882)
print(d3)
#{'Nombre': 'Sara', 'Edad': 27, 'Documento': 1003882}
```

Algunas propiedades de los diccionario en Python son las siguientes:

Son dinmicos, pueden crecer o decrecer, se pueden aadir o eliminar elementos.
Son indexados, los elementos del diccionario son accesibles a travs del key.
Y son anidados, un diccionario puede contener a otro diccionario en su campo value.

```python
for n in range(2, 120):
    for x in range(2, n):
        if n % x == 0:
            print(n, 'es igual a', x, '*', n//x)
            break
    else:
        # el blucle o loop falla cuando no encuentra un rango
        print(n, 'es un numero primo')
```

Por otro lado, las sentencias **continue** (tambin tomadas de C) continan con la siguiente iteracin del bucle:

```python
for num in range(2, 10):
    if num % 2 == 0:
        print("Numero Par", num)
        continue
    print("Numero Impar", num)
```

# Bucle while

Una declaracin de bucle while en el lenguaje de programacin Python ejecuta repetidamente una declaracin de destino siempre que una condicin dada sea verdadera.<br>
Cuando la condicin se vuelve falsa, el control del programa pasa a la lnea que sigue inmediatamente al ciclo.<br>
Creemos una miniserie de Fibonacci para comprender cmo funcionan los bucles While:

```python
a, b = 0, 1
while b < 10:
    print(b)
    a, b = b, a+b
```

```python
value = True
while value:
    print('Todo esta bien!')
    break
    
```
